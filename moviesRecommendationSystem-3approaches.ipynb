### Calculate a weighted rating

WR = (v / (v+m)) * R + (m / (v+m)) * C

v - number of votes for a movie

m - minimum number of votes required

R - average rating of the movie

C - average rating across all movies

```python
m = movies["vote_count"].quantile(0.92)
m
```

```python
C = movies["vote_average"].mean()
C
```

```python
movies_filtered = movies.copy().loc[movies["vote_count"] >= m]
movies_filtered
```

```python
def weighted_rating(df, m=m, C=C):
  R = df["vote_average"]
  v = df["vote_count"]
  wr =((v / (v+m)) * R) + ((m / (v+m))* C)
  return wr
```

```python
movies_filtered["weighted_rating"] = movies_filtered.apply(weighted_rating, axis=1)
movies_filtered
```

```python
movies_filtered.sort_values("weighted_rating", ascending=False)[["title","weighted_rating"]].head(10)
```

# Content-based Filtering




```python
movies1 =  pandas.read_csv("movies.csv")
```

```python
movies1
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words="english")
```

```python
movies["overview"] = movies["overview"].fillna("")
```

```python
tfidf_matrix = tfidf.fit_transform(movies["overview"])
```

```python
pandas.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

```python
tfidf_matrix.shape
```

## Similarity Matrix

```python
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix = linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
similarity_matrix
```

```python
similarity_matrix.shape
```

### Find similar movies for given movie

```python
def similar_movie(movie_title, nr_movies):
  idx = movies.loc[movies["title"]==movie_title].index[0]
  scores = list(enumerate(similarity_matrix[idx]))
  scores = sorted(scores, key=lambda x:x[1], reverse = True)
  movies_indices=  [tpl[0] for tpl in scores[1:nr_movies+1]]
  similar_titles = list(movies["title"].iloc[movies_indices])
  return similar_titles
```

```python
similar_movie("Kung Fu Panda 3", 5)
```

```python
from surprise import Dataset, Reader

reader = Reader(rating_scale=(1,5))
dataset = Dataset.load_from_df(ratings1, reader)
dataset
```

```python
trainset = dataset.build_full_trainset()
trainset
```

```python
from surprise import SVD

model = SVD()
```

```python
model.fit(trainset)
```

```python
model.predict(9,1989)
```

```python
from surprise import model_selection

model_selection.cross_validate(model, dataset, measures=['RMSE','MAE'])
```
